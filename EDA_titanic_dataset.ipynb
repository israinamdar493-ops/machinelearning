Question 3: Compute the basic descriptive statistics (mean, median, minimum, maximum, standard deviation) for numerical columns.

```python
# Mean
print("Mean")
print(df.mean(numeric_only=True))

print("\nMedian")
print(df.median(numeric_only=True))

print("\nMinimum")
print(df.min(numeric_only=True))

print("\nMaximum")
print(df.max(numeric_only=True))

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

```python
print(df.describe())
```

The mean gives the average value of each numerical column.



The median gives the middle value of the data when arranged in ascending order.



The minimum and maximum values show the range of the data.



The standard deviation indicates how much the values are spread around the average.



The describe() function provides all the important descriptive statistics in a single table.



These statistics give a quick summary of the numerical columns and help understand the overall distribution of the data before further analysis.

Question 6: Extract rows where a numerical column value is greater than a specific threshold.

 The dataset contains 64 passengers whose age is greater than 50 years.



Only records satisfying the specified condition are displayed.



Passengers with an age of 50 or less are excluded from the output.



Missing age values (NaN) are automatically excluded during filtering.



Filtering data using conditions helps focus on a specific subset of the dataset for analysis.

Question 7: Sort the dataset based on a numerical column in descending order.

Passengers who paid the highest ticket fare appear at the top of the dataset.



The sort_values() function is used to sort data based on any selected column.



Setting ascending=False arranges the values from highest to lowest.



The head() function helps display only the first few sorted records instead of the entire dataset.

Question 8 : Visualize the distribution of a categorical column.

```python
df["Sex"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.show()
```

The dataset contains two categories in the Sex column: Male and Female.



The number of Male passengers is greater than the number of Female passengers.



A bar chart makes it easy to compare the frequency of different categories.



The value_counts() function calculates the frequency of each category.



The plot(kind="bar") function is useful for visualizing categorical data.

Question 9: Plot a histogram of a numerical column to visualize its distribution.

```python
plt.hist(df["Age"].dropna(), bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

The hist() function creates a histogram for the Age column.



The dropna() function removes missing age values (NaN) before plotting the graph.



The parameter bins=20 divides the age values into 20 intervals (groups).



The x-axis represents the age groups, while the y-axis represents the number of passengers in each age group.



Taller bars indicate age groups with a larger number of passengers, while shorter bars indicate fewer passengers.



The histogram helps visualize how the numerical data is distributed.

Question 10: Visualize the relationship between two numerical columns using a scatter plot.

```python
plt.scatter(df["Age"], df["Fare"])

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

Each dot in the scatter plot represents one passenger.



The x-axis represents Age, and the y-axis represents Fare.



Passengers of similar ages have paid different ticket fares.



No clear relationship is observed between passengers' age and the fare they paid.



Scatter plots are useful for comparing two numerical variables and identifying patterns, trends, or outliers in the data.

Question 11: Visualize pairwise relationships in a subset of features.

```python
import seaborn as sns

sns.pairplot(df[["Age", "Fare", "SibSp", "Parch"]])

plt.show()
```

The pair plot displays both the distribution of individual features and the relationship between pairs of features.



Most passengers were between 20 and 40 years of age.



Most passengers paid lower ticket fares, while only a few paid very high fares.



Most passengers travelled with 0 or 1 sibling/spouse and 0 parents/children.



No strong relationship is observed between Age and Fare.



A few outliers are present, especially in the Fare column.



The pair plot is useful for identifying distributions, patterns, relationships, and outliers before performing further data analysis or machine learning.
