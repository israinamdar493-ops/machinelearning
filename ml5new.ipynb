```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Generate 100 random x values in [0, 1]
np.random.seed(0)  # to keep results consistent
X = np.random.rand(100, 1)

# Step 2: Label the first 50 points
# Class1 = 0 (for xi < 0.5), Class2 = 1 (for xi > 0.5)
y_train = np.array([0 if x < 0.5 else 1 for x in X[:50]])

# Step 3: Prepare data
X_train = X[:50]
X_test = X[50:]

# Step 4: Define k values to try
k_values = [1, 2, 3, 4, 5, 20, 30]

# Step 5: Run k-NN and classify X[51] to X[100]
print("k-NN Classification of X[51] to X[100]:")
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(f"Results for k = {k}:")
    for i, (x_val, label) in enumerate(zip(X_test, y_pred), start=51):
        class_name = "Class1" if label == 0 else "Class2"
        print(f"x[{i}] = {x_val[0]:.4f} - Predicted: {class_name}")
    print("-" * 50)
    
    
    
    
```
